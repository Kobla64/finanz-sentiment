```python
import pandas as pd
import matplotlib.pyplot as plt

# Beim Einlesen direkt sagen, dass die erste Spalte (0) das Datum ist
df = pd.read_csv("data/raw_stock_data.csv", index_col=0, parse_dates=True)

# Jetzt ist das Datum der Index und bereits im richtigen Zeitformat
df['Close'].plot(figsize=(10, 5), title="Kursverlauf 2024")

plt.grid()
plt.show()
```
